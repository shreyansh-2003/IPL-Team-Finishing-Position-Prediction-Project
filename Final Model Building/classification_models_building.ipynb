### Finding the accuracy for each of the classification algorithm: 

#### Hypothesis:
Winning or success rate in a team sport like cricket, football does not entirely depend on the individual performance of the player, rather a team wins or performs well when most of the players of that team plays exceptionally good.

````python
# Importing libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, f1_score, classification_report

# Split the data into X and y
X = df_new.iloc[:, :-1].values
y = df_new.iloc[:, -1].values

# Split the data into train and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Decision Tree Classifier
dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_train, y_train)
dt_pred = dt.predict(X_test)
dt_acc = accuracy_score(y_test, dt_pred)
print("Decision Tree Classifier Accuracy: {:.2f}%".format(dt_acc*100))

# K-Nearest Neighbors Classifier
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
knn_pred = knn.predict(X_test)
knn_acc = accuracy_score(y_test, knn_pred)
print("K-Nearest Neighbors Classifier Accuracy: {:.2f}%".format(knn_acc*100))

# Naive Bayes Classifier
nb = GaussianNB()
nb.fit(X_train, y_train)
nb_pred = nb.predict(X_test)
nb_acc = accuracy_score(y_test, nb_pred)
print("Naive Bayes Classifier Accuracy: {:.2f}%".format(nb_acc*100))

# Logistic Regression Classifier
lr = LogisticRegression(random_state=42)
lr.fit(X_train, y_train)
lr_pred = lr.predict(X_test)
lr_acc = accuracy_score(y_test, lr_pred)
print("Logistic Regression Classifier Accuracy: {:.2f}%".format(lr_acc*100))

# Random Forest Classifier
rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, y_train)
rf_pred = rf.predict(X_test)
rf_acc = accuracy_score(y_test, rf_pred)
print("Random Forest Classifier Accuracy: {:.2f}%".format(rf_acc*100))

# SVM Classifier
svm = SVC(random_state=42)
svm.fit(X_train, y_train)
svm_pred = svm.predict(X_test)
svm_acc = accuracy_score(y_test, svm_pred)
print("SVM Classifier Accuracy: {:.2f}%".format(svm_acc*100))
````

#### Observation:
- As I plotted the ROC courve for each of the algorithm seperately I noticed that KNN is overfitting the values and so the area under its curve is the highest so we will not consider KNN as the best algorithm for our model instead we will go for Random Forest as it is the second model having the highest area under its curve and might be a better option as it has a relatively high accuracy and is not overfitting the data.
- Even though the accuracy for Random forest (65.47%) is not that high it still makes our hypothesis true that in a team sport like cricket, team performance does not entirely depend on the individual performance of the players.

## ROC Curve

````python
from sklearn.metrics import roc_auc_score, roc_curve
import matplotlib.pyplot as plt

# Define the list of algorithms to compare
algorithms = ['KNN', 'Logistic Regression', 'Decision Tree', 'Random Forest', 'SVM', 'Naïve Bayes']

# Define the list of target combinations to compare
targets = [7.0, 2.0, 3.0, 5.0, 8.0, 6.0, 4.0, 1.0, 9.0, 0.0]

# Create a new figure with multiple subplots
fig, axs = plt.subplots(nrows=2, ncols=3, figsize=(20, 10))

# Iterate over the algorithms
for i, alg in enumerate(algorithms):
    # Set the current subplot
    ax = axs[i//3][i%3]
    
    # Set the subplot title and axis labels
    ax.set_title(alg)
    ax.set_xlabel('False Positive Rate')
    ax.set_ylabel('True Positive Rate')
    ax.set_xlim([0.0, 1.0])
    ax.set_ylim([0.0, 1.05])
    
    # Iterate over the target combinations
    for target in targets:
        # Get the predictions and probabilities for this target using the current algorithm
                # Get the predictions and probabilities for this target using the current algorithm
        if alg == 'KNN':
            knn = KNeighborsClassifier(metric='euclidean', n_neighbors=3, weights='distance')
            knn.fit(X_train, y_train == target)
            y_pred = knn.predict(X_test)
            y_proba = knn.predict_proba(X_test)
        elif alg == 'Logistic Regression':
            logreg = LogisticRegression(C=0.01, penalty='l2')
            logreg.fit(X_train, y_train == target)
            y_pred = logreg.predict(X_test)
            y_proba = logreg.predict_proba(X_test)
        elif alg == 'Decision Tree':
            dt = DecisionTreeClassifier(max_depth=15, min_samples_leaf=1, min_samples_split=5)
            dt.fit(X_train, y_train == target)
            y_pred = dt.predict(X_test)
            y_proba = dt.predict_proba(X_test)
        elif alg == 'Random Forest':
            rf = RandomForestClassifier(max_depth=15, min_samples_leaf=1, min_samples_split=5, n_estimators=100)
            rf.fit(X_train, y_train == target)
            y_pred = rf.predict(X_test)
            y_proba = rf.predict_proba(X_test)
        elif alg == 'SVM':
            svm = SVC(probability=True)
            svm.fit(X_train, y_train == target)
            y_pred = svm.predict(X_test)
            y_proba = svm.predict_proba(X_test)
        elif alg == 'Naïve Bayes':
            nb = GaussianNB()
            nb.fit(X_train, y_train == target)
            y_pred = nb.predict(X_test)
            y_proba = nb.predict_proba(X_test)
        
        
        # Compute the AUC-ROC score for this target using the current algorithm
        auc_score = roc_auc_score(y_test == target, y_proba[:, 1])
        
        # Compute the false positive rate, true positive rate, and thresholds for this target using the current algorithm
        fpr, tpr, thresholds = roc_curve(y_test == target, y_proba[:, 1])
        
        # Plot the ROC curve for this target using the current algorithm
        ax.plot(fpr, tpr, label='Target = %s (AUC = %0.2f)' % (target, auc_score))
        
    

plt.legend()
plt.show()
````

#### Observation:
As it is clearly depicted in the plots that KNN is overfitting and is catching all the different classes too well, hence we will chose the next best algorithm, Random Forest.
